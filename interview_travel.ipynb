### 1.2. A/B Testing

The number of visitors on your website equals the number of opportunities you have to expand your business by acquiring new customers and build relationships by catering to existing ones. And it is your conversion funnel that decides whether your website gets good traffic and if it converts more visitors. Businesses want visitors to take action (also called a conversion) on their website, and the rate at which a site can drive this is called its “conversion rate.” The more optimized your funnel, the higher is the visitors’ chance to convert. [Refrence](https://vwo.com/ab-testing/?utm_campaign=tof_ugc)

One way to optimize your website’s funnel is **A/B testing**. A/B testing (also sometimes referred to as split testing) is the practice of showing two variants of the same web page to different segments of visitors at the same time and comparing which variant drives more conversions. Typically, the one that gives higher conversions is the winning variant, applying, which can help you optimize your site for better results. [Refrence](https://vwo.com/ab-testing/?utm_campaign=tof_ugc)

Read more at: [VWO](https://vwo.com/ab-testing/?utm_campaign=tof_ugc)

### 1.3. About Data

We have run two A/B tests (testing different features, one in February, the other in May) with a classical setup: show a control group to the majority of our users (constituting a subset of website visits), and a test group (containing the new feature) to the remaining users (and corresponding remaining visits).
You are provided with one csv file for each of these two tests, containing the results collected in terms of our main base metrics.

* `ymd` - the date
* `platform` - the country-specific XXXXXXXXXX platform 
* `group` - control or test
* `visits` - how many website visits from our users were logged
* `clicks` - how many times our users clicked in one of our prices and were re-directed to
the website of one of our advertisers. Note that each individual visit may or may not
include clicks.
* `revenue` - how much money (in EUR) was generated for XXXXXXXXXX. You can work under
the assumption that this revenue is generated under a CPC model.
* `bookings` - how many accommodation reservations were made in our advertisers’
websites.
* `booking_amount` - the corresponding amount (in EUR) paid by the users for the
accommodation reservations.

### 2.3. The global and platform average revenue per click

The revenue per click can be calculated as the sum of revenue divided by the sum of clicks in a given time. The global RPC is as followed:

```python
# sum of revenue devided by sum of clicls
global_RPC= test1.sum().revenue/test1.sum().clicks
print('Global RPC:',global_RPC)
```

Same as global RPC, we can calculate RPC for each platform:

```python
## for test1 dataset
# sum of revenue for each platform
df_temp_1= test1.groupby('platform').sum()
# sum of revenue devided by sum of clicls for each platform as a new column
df_temp_1['revenue_per_click_Feb']=df_temp_1['revenue']/df_temp_1['clicks']
# sort data
df_temp_1.sort_values('revenue_per_click_Feb',ascending=False, inplace=True)

## for test2 dataset
# sum of revenue for each platform
df_temp_2= test2.groupby('platform').sum()
# sum of revenue devided by sum of clicls for each platform as a new column
df_temp_2['revenue_per_click_May']=df_temp_2['revenue']/df_temp_2['clicks']
# sort data
df_temp_2.sort_values('revenue_per_click_May',ascending=False, inplace=True)

# concat data and keep only rpc columns
df_temp= pd.concat([df_temp_1['revenue_per_click_Feb'],df_temp_2['revenue_per_click_May']], axis=1)
# check data
df_temp
```

The chart below shows the average revenue per click for each platform for each given dataset.

```python
# set background
sns.set_style('darkgrid')
# create a figure, set size and dpi
fig, ax = plt.subplots(figsize=(10,5),dpi=100)
# draw a line for Global RPC value
plt.axhline(y=global_RPC, xmin=0.01, xmax=0.99, color='k', ls='--')
# write text above the line
plt.text(2.5, 1, "Global RPC: {:.2f}".format(global_RPC), fontsize=15, fontweight='bold')
# plt bar chart
labels = df_temp.index
x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars
rects1 = ax.bar(x - width/2, df_temp['revenue_per_click_Feb'], width, label='February')
rects2 = ax.bar(x + width/2, df_temp['revenue_per_click_May'], width, label='May')
ax.set_xticks(x)
ax.set_xticklabels(labels)
# set title and style xy ticks and labels
plt.title('\nPlatform Average Revenue per Click (EUR/Click)\n', fontsize=22, fontweight='bold')
plt.xlabel('Platform', fontsize=16)
plt.xticks(fontsize=12, rotation=0)
plt.ylabel('Average RPC', fontsize=16)
plt.yticks(fontsize=12)
plt.legend(handles=[rects1,rects2])
# show plot
plt.show()
```

### 2.5. Relation between base metrics and valuable KPIs

First of all, let's explore all the base metrics based on the given data. The three binary metrics in the datasets, respectivly **visits**, **clicks**, and **bookings** can explain the typical user flow.

In this case, the user flow starts by a specific number of users visiting the website. Then a subset of those user might click on the button. Finally, a smaller groupe of them make a booking. This flow, which also sometimes is called customer funnel, can be seen in below diagram. Although this is a very simplistic idea and there is a lot of back and forth swirl between the different states, it can give us good insight about the process and help us in metric choices (which we will discuss in next chapter).

Based on this funnel, we can define three different rates or three different probabilities.
* rate of clicks to visits (or Click-Through-Probability)
* rate of booking to visits (or probability that visitors make a booking)
* rate of booking to clicks (or probability that people who clicked make a booking)

```python
init_notebook_mode(connected=True)
cf.go_offline()
# calculate click share and booking share regarding the visits
clicks_share= str(int(test1.sum().clicks/test1.sum().visits*100))
bookings_share= str(int(test1.sum().bookings/test1.sum().visits*100))
# create a figure and plot a funnel with three blocks
fig = go.Figure(go.Funnelarea(
      values = [3,2,1], 
      text = ["Visits (100%)","Clicks ("+clicks_share+'%)', "Bookings ("+bookings_share+'%)'],
      marker = {"colors": ["silver","lightblue", "lightsalmon",],
                "line": {"color": ["wheat", "wheat","wheat"], 
                         "width": [5,5,5]}},
      textfont = {"family": "Old Standard TT, serif", "size": 16, "color": "black"}, 
      opacity = 0.7, textinfo = 'text'),
      layout={'showlegend':False})
# set title
fig.update_layout(
    title={
        'text': "User Flow Funnel",
        'y':0.93,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    font=dict(
        size=20,
        color="black"))
# show plot
fig.show()
```

In the above funnel which is drawn for the test1 dataset, it can be seen that 90% of the visitors click, and only 4% of the make a booking.

### 3.1. Chose a metric to evaluate test 1

It would have been very helpful to have some information about the tests for the analysis. The metric choice is usually done in the design step before running a test. So one would know which metrics are supposed to change and which not. Anyway, we try to do a meaningful analysis. First of all let's take a look at the metrics for the different groups:

```python
# store each group in a seperate dataframe
test_df_1= test1[test1['group']=='test']
control_df_1= test1[test1['group']=='control']
# sum the columns and concat dataframes
test1_sum= pd.concat([test_df_1.sum(numeric_only=True).to_frame('test').T,
                control_df_1.sum(numeric_only=True).to_frame('control').T,
                test1.sum(numeric_only=True).to_frame('total').T])
# build new columns fo the metrics (rates. probabilities, etc)
test1_sum['clicks/visits']=test1_sum['clicks']/test1_sum['visits']
test1_sum['revenue/visits']=test1_sum['revenue']/test1_sum['visits']
test1_sum['bookings/visits']=test1_sum['bookings']/test1_sum['visits']
test1_sum['booking_amount/visits']=test1_sum['booking_amount']/test1_sum['visits']
test1_sum['booking_amount/booking']=test1_sum['booking_amount']/test1_sum['bookings']
```

### 3.2. Test 1 performance summary and recommendations

First of all, we evaluate the performance of the new feature on all the data (all platforms).

So our hypothesises are:

* **Null Hypothesis**: The control and test group have the same probability to book

$$H_0 : P_{cont} - P_{test} = 0$$


* **Alternative Hypothesis**: The control and test group have a different probability to book

$$H_A : P_{cont} - P_{test} = d$$

```python
# print probability to book for each group
p_cont= test1_sum.loc['control','bookings/visits']
p_test= test1_sum.loc['test','bookings/visits']
print('P_cont: ', p_cont)
print('P_test: ', p_test)
```

We should see how likely it is that this change in the probability to book is by chance. Since we have a large number of data points we can define a low practical significance boundray. I used this [online calculator](https://www.evanmiller.org/ab-testing/sample-size.html) to specify $d_{min}$.

$d_{min}=0.001$

Other chosen metrics:
* confidence interval 95% (significance level $\alpha$ 0.05)
* statistical power 90% ($\beta$ 0.1)

First we estimate pool probability $\hat{P}_{pool}$:


$$\hat{P}_{pool} = \frac{Bookings_{test}+Bookings_{control}}{Visits_{test}+Visits_{control}} $$

```python
p_pool= test1_sum.loc['total','bookings/visits']
```


Then we estimate standard error. We can assume metrics which are probabilities ($\hat{p}$) are binomially distributed, so we can use this formula for the standard deviation: 

$$SE = \sqrt{\hat{p}*(1-\hat{p})*  (\frac{1}{Visits_{control}}+\frac{1}{Visits_{test}})    }$$

```python
SE_pool= np.sqrt(p_pool*(1-p_pool)*((1/test1_sum.loc['control','visits'])+(1/test1_sum.loc['test','visits'])))
```

Difference in probability in test and control group:

```python
d_hat= p_test-p_cont
```

Confidence Interval:

```python
d_hat_interval= [d_hat-(1.96*SE_pool),d_hat+(1.96*SE_pool)]
```

```python
practical_significance_boundray= [-0.001,0.001]
```

```python
print('d_hat: ',"{:.4f}".format(d_hat))
print('confidence interval: ', "[{:.4f}, ".format(d_hat_interval[0]), "{:.4f}]".format(d_hat_interval[1]))
print('practical significance boundray: ', practical_significance_boundray)
print()
```

```python
# set background to white
sns.set_style('white')
# create figure
plt.figure(figsize=(10,3),dpi=100)
# set title
plt.title("\nConfidence Interval \nfor the Difference in Booking Prabability\n", 
          fontsize=22, fontweight='bold')
# plot confidence interval
plt.plot([d_hat_interval[0],d_hat,d_hat_interval[1]],3*[0.5],'-o',markersize=10)
# change color of span to gray
plt.axvspan(-0.001,0.001, facecolor='gray', alpha=0.2)
# draw three --- lines
plt.axvline(x=-0.001, c='k',lw=2, ls='--')
plt.axvline(x=0, c='k', lw=2, ls='--')
plt.axvline(x=0.001, c='k', lw=2, ls='--')
# style the graph
plt.xlim([-0.003,0.003])
plt.xticks(ticks=[-0.001,0,0.001], labels=['$-d_{min}=-0.001$',0,'$d_{min}=0.001$'])
plt.yticks([])
plt.show()
```

The point estimate $d_{hat}$ and both ends of the confidence interval are lower than the practical significance boundary. Based on this, it is highly probable that the probability of booking did, in fact, change more than the practical significance boundary.

Let's look at revenue per visits and booking_amount per booking and see how they differ in the test and control gorups:

As can be seen in the plots, JP has a higher booking rate in test group which is interesting to investigate. We calculate $d_{min}$ and the confidence interval for each platform.

Because the test size is smaller, we need new assumptions this time:\
Statistical power: 0.8\
Significance level: 0.05\
$d_{min}=0.0015$

First, we look at JP specifically.

```python
# set probabilities for JP
p_cont_JP= test1_sum_platforms.iloc[7,8]
p_test_JP= test1_sum_platforms.iloc[6,8]
p_pool_JP= (test1_sum_platforms.iloc[6,4]+test1_sum_platforms.iloc[7,4])/\
           (test1_sum_platforms.iloc[6,1]+test1_sum_platforms.iloc[7,1])
# calculate pool standard error for JP 
SE_pool_JP= np.sqrt(p_pool_JP*(1-p_pool_JP)*
                    ((1/test1_sum_platforms.iloc[6,1])+
                     (1/test1_sum_platforms.iloc[7,1])))
# point estimate
d_hat_JP= p_test_JP-p_cont_JP
# confidence interval
d_hat_interval_JP= [d_hat_JP-(1.96*SE_pool_JP),d_hat_JP+(1.96*SE_pool_JP)]
# practical significance boundray
practical_significance_boundray_JP= [-0.0015,0.0015]
# print results
print('\nd_hat_JP: ',"{:.4f}".format(d_hat_JP))
print('confidence interval JP: ', "[{:.4f}, ".format(d_hat_interval_JP[0]), "{:.4f}]".format(d_hat_interval_JP[1]))
print('practical significance boundray JP: ', practical_significance_boundray_JP)
print()
```

In this case, our point estimate is lower than the practical significance boundary. The confidence interval is almost within the boundaries who implies an almost neutral change. We cannot conclude that the feature has a positive impact on the probability of booking.

There is a slight increase in revenue per click and booking_amount per booking. We can investigate these metrics statistically further and see if it is practically significant. (To keep this short, I do not go through all the further analysis)

Let's calculate the confidence interval for all the platforms and visualize to gain a better insight.

```python
# create an empty df to store confidence intervals in
df_d_hat=pd.DataFrame(index=['low_boundray','d_hat','high_boundray'])
# loop through each platform
for i in range(0,12,2):
    # set probabilities
    p_test_temp= test1_sum_platforms.iloc[i,8]    
    p_cont_temp= test1_sum_platforms.iloc[i+1,8]
    p_pool_temp= (test1_sum_platforms.iloc[i,4]+test1_sum_platforms.iloc[i+1,4])/\
               (test1_sum_platforms.iloc[i,1]+test1_sum_platforms.iloc[i+1,1])
    # calculate pool standard error
    SE_pool_temp= np.sqrt(p_pool_temp*(1-p_pool_temp)*
                        ((1/test1_sum_platforms.iloc[i,1])+
                         (1/test1_sum_platforms.iloc[i+1,1])))
    # point estimate
    d_hat_temp= p_test_temp-p_cont_temp
    # store in df
    df_d_hat[test1_sum_platforms.index[i]]=[d_hat_temp-(1.96*SE_pool_temp),
                                            d_hat_temp,
                                            d_hat_temp+(1.96*SE_pool_temp)]
```
